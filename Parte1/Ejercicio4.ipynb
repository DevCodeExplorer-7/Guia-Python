Programar en python la conjetura de Collatz y grafique los valores de la función con respecto a la
iteraciones.

```python
import matplotlib.pyplot as plt  #biblioteca de gráficos
```

```python

def secuencia_collatz(n, sequence=None):
    """
    Calcula la secuencia de Collatz de forma recursiva para un número entero n.

    :param n: Número inicial de la secuencia.
    :param sequence: Lista que almacena la secuencia (opcional, se crea si es None).
    :return: Lista que contiene la secuencia de Collatz.
    """
    if sequence is None:
        sequence = [n]
    if n == 1:
        return sequence
    elif n % 2 == 0:
        sequence.append(n // 2)
        return secuencia_collatz(n // 2, sequence)
    else:
        sequence.append(3 * n + 1)
        return secuencia_collatz(3 * n + 1, sequence)


def secuencia_collatz_plot(n, max_iterations=1000):
    """
    Genera un gráfico de la secuencia de Collatz para un número entero n.

    :param n: Número inicial de la secuencia.
    :param max_iterations: Número máximo de iteraciones (predeterminado: 1000).
    """
    sequence = secuencia_collatz(n)
    iterations = list(range(len(sequence)))

    # Estilo del gráfico
    plt.figure(figsize=(10, 6))  # Tamaño de la figura
    plt.plot(iterations, sequence, marker='o', linestyle='-', markersize=3, color='b')  # Estilo de línea y marcadores
    plt.xlabel('Iteraciones', fontsize=12)  # Etiqueta del eje X
    plt.ylabel('Valor de la Secuencia de Collatz', fontsize=12)  # Etiqueta del eje Y
    plt.title(f'Conjetura de Collatz para n = {n}', fontsize=14)  # Título del gráfico
    plt.grid(True, linestyle='--', alpha=0.6)  # Rejilla punteada
    print()# salto de linea para los graficos
    plt.show()

# Ejemplos
valor_numero = [63, 23, 2023]

for n in valor_numero:
    secuencia_collatz_plot(n)
```
